```python
"""
Create a sample set
"""

import textwrap

fname='test.txt'
with open(fname,'w') as f:
    f.write(textwrap.dedent('''
    1,5
    2,3
    3.5,3
    4,5
    5,4
    6,3
    7,5
    '''))
```

```python
"""
Use dataset from file
"""

import numpy as np
import matplotlib.pyplot as plt

fname='test.txt'
x, y = np.loadtxt(fname, delimiter=',', unpack=True)
# plt.plot(x,y, label='Loaded from file!', marker='o', linestyle = '-')
plt.plot(x,y, '.-', label='Loaded from file!')
plt.legend()
# plt.ylim(2,6)   # or xlim  
plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(True)
plt.title('Graph of data - Check it out')
# plt.text(2, 5., "Hallo")
# plt.show()
plt.savefig('test2.jpg')  
```

```python
"""
Use dataset from file
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

fname='test.txt'
x, y = np.loadtxt(fname, delimiter=',', unpack=True)

f1 = interp1d(x, y)  # linear
f2 = interp1d(x, y, kind='cubic')
xnew = np.linspace(min(x), max(x),num=100,  endpoint=True)  

# plt.plot(x, y, 'o', xnew, f1(xnew), '-', xnew, f2(xnew), '--')
plt.plot(x, y, 'o',  xnew, f2(xnew), '--')
# plt.legend(['data', 'linear', 'cubic'], loc='best')

# plt.ylim(2,6)   # or xlim  
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Graph of data - Check it out')
plt.grid(True)
# plt.text(2, 5., "Hallo")
# plt.show()
plt.savefig('test2.jpg')  
```
