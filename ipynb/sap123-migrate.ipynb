```python
import os
import sqlite3
import csv
import unicodedata

DIR_NAME = 'C:\\data\\python\\sap123-convert'
DIR_POSTS = 'C:\\data\\github\\sap123\\a'
DB = os.path.join(DIR_NAME, 'sap123-1.db')
```

```python
def blog_post(rows):
    counter = 0
    con = sqlite3.connect(DB)
    cur = con.cursor()
    cur.execute("""DROP TABLE IF EXISTS blog_post""")
    cur.execute("""CREATE TABLE blog_post (
    post, section, priority, title, slug, f1, contents, keywords,
    name1, name2, date_create, date_edit, t, f2, tel, active)""" )    
    for row in rows:
        fields = row.split("\t")
        if len(fields) > 15:
            cur.execute("""INSERT INTO blog_post (
            post, section, priority, title, slug, f1, contents,keywords,
            name1, name2, date_create, date_edit, t,f2,tel, active
            ) VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?);""", fields[:16])
            counter += 1
    con.commit()
    con.close()
    print("blog_post", counter, " records with ",16," fields each")
    
def b_post(rows):
    counter = 0
    con = sqlite3.connect(DB)
    cur = con.cursor()
    cur.execute("""DROP TABLE IF EXISTS b_post""")
    cur.execute("""CREATE TABLE b_post (
    post, section, priority, title, slug, f1, contents, keywords,
    name1, name2, date_create, date_edit, t, f2, tel, active)""" )    
    for row in rows:
        fields = row.split("\t")
        if len(fields) > 10:
            cur.execute("""INSERT INTO b_post (
            post, section, priority, title, slug, f1, contents,keywords,
            name1, name2, date_create, date_edit,t, f2, tel
            ) VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?);""", fields[:15])
            counter += 1
    con.commit()
    con.close()
    print("b_post", counter, " records with ",15," fields each")
    
def blog_section(rows):
    counter = 0
    con = sqlite3.connect(DB)
    cur = con.cursor()
    cur.execute("""DROP TABLE IF EXISTS blog_section""")
    cur.execute("""CREATE TABLE blog_section (
    a, b, c, data, f, active 
    )""" )    
    for row in rows:
        fields = row.split("\t")
        if len(fields) >= 6:
            cur.execute("""INSERT INTO blog_section (
            a, b, c, data, f, active 
            ) VALUES (?,?,?,?,?,?);""", fields[:6])
            counter += 1
    con.commit()
    con.close()
    print("blog_section", counter, " records with ",6," fields each")

def b_section(rows):
    counter = 0
    con = sqlite3.connect(DB)
    cur = con.cursor()
    cur.execute("""DROP TABLE IF EXISTS b_section""")
    cur.execute("""CREATE TABLE b_section (
    a, b, c, data, f, active 
    )""" )    
    for row in rows:
        fields = row.split("\t")
        if len(fields) >= 6:
            cur.execute("""INSERT INTO b_section (
            a, b, c, data, f, active 
            ) VALUES (?,?,?,?,?,?);""", fields[:6])
            counter += 1
    con.commit()
    con.close()
    print("b_section", counter, " records with ",6," fields each")
    

def readfile(my_file):
    df = os.path.join(DIR_NAME, my_file)
    with open(df, newline='', encoding = 'ISO-8859-1') as f:
        raw_data = f.read()
        rows = raw_data.split("\n") 
    if my_file == 'blog_post.txt':
        blog_post(rows)
    elif my_file == 'blog_section.txt':
        blog_section(rows)
    elif my_file == 'b_post.txt':
        b_post(rows)
    elif my_file == 'b_section.txt':
        b_section(rows)

def get_files ():
    entries = os.listdir(DIR_NAME)
    for f in entries:
        readfile(f) 
get_files()
print('done ...')
```
